# Transactions by Channel (Online, Walk-in, etc) and Park

In this particular EDA, we want to analyze the transactions by channel and park (Sunway Lagoon VS Lost World). We would want to see if there is any particular trend, eg, more individuals are paying using online in Sunway Lagoon, etc. 

## List of variables

1. Environment: Specify whether the environment is Sunway Lagoon or Lost World
- WHEN Environment = 1 THEN 'Sunway Lagoon'
- WHEN Environment = 2 THEN 'Lost World'

2. Transaction data
- Complete_QTY: Total Visitors
- OriginalAmount * Complete_QTY: Total Revenue

3. Park: Revenue Grouping (whether you are online, walk-in, bulk buying, etc)
- We should understand the different unique values available

4. TranDate: Transaction Date
- According to our EDA, we only focus on the data from 2023 to 2024 to train our model, skipping earlier years that is affected by COVID-19.

## Loading Dataset

```python
# Connect to the database
con = duckdb.connect("theme_park_data.db")

# Extract the necessary information for EDA
# Transactions data, channels (revenue grouping), and type of park/environment
# Focus from 2023 to 2024

df_channel_park = con.execute("""
SELECT
    TRIM(Park) as Grouping,
    CASE
        WHEN Environment = 1 THEN 'Sunway Lagoon'
        WHEN Environment = 2 THEN 'Lost World'
    END AS Park,
    OriginalAmount AS Unit_Price,
    Complete_QTY AS No_of_Visitors,
    (OriginalAmount * Complete_QTY) AS Total_Revenue
FROM theme_park_database
WHERE Environment IN (1, 2) 
    AND Grouping IS NOT NULL
    AND TranDate >= TIMESTAMP '2023-01-01 00:00:00'
    AND TranDate < TIMESTAMP '2025-01-01 00:00:00'
""").df()

df_channel_park.head()

# Clean the 'Grouping' column directly in the dataframe
# .strip() removes spaces, .title() converts 'walk in' to 'Walk In'
df_channel_park['Grouping'] = df_channel_park['Grouping'].str.strip().str.title()
```

### 🟡 Dirty Data Processing 

Hence, here are the steps I would do to temporary "correct" the dataset before heading into generating and comparing the transactions. 
1. Correcting the sign of the ticket price (I suspect that is a typo of negative sign, and I believe that there should be a separate section by the data regarding reimbursement column)
2. Correcting the sign of the number of visitors (because number of visitor is never negative).
3. Dropping all the number of visitors which is <1, to ensure our result is still interpretable and not having weird visitor number. 

```python
# 1. Correcting the sign of the ticket price (OriginalAmount)
# Using absolute value to ensure all prices are positive
df_channel_park['Unit_Price'] = df_channel_park['Unit_Price'].abs()

# 2. Correcting the sign of the number of visitors (Complete_QTY)
# Ensuring that negative visitor counts (likely reversal logs) are treated as positive volume
df_channel_park['No_of_Visitors'] = df_channel_park['No_of_Visitors'].abs()

# 3. Drop fractional visitors but keep whole numbers (including 0)
# This keeps 0.0, 1.0, 2.0 etc., and removes 0.5, 1.7, etc.
initial_count = len(df_channel_park)
df_channel_park = df_channel_park[df_channel_park['No_of_Visitors'] % 1 == 0]

# Recalculate total_revenue to ensure it matches the corrected signs
df_channel_park['Total_Revenue'] = df_channel_park['Unit_Price'] * df_channel_park['No_of_Visitors']

dropped_count = initial_count - len(df_channel_park)

print(f"Data Cleaning Complete:")
print(f"- Fixed negative signs for Prices and Visitor counts.")
print(f"- Dropped {dropped_count} rows with visitor counts < 1.")
print(f"- New dataframe shape: {df_channel_park.shape}")
print(f"Amount of data lost: {dropped_count} rows ({(dropped_count/initial_count)*100:.2f}%)")

print("### Summary Statistics by Park and Channel ###")
summary_stats = df_channel_park.groupby(['Park', 'Grouping']).agg({
    'Total_Revenue': ['sum', 'mean', 'std'],
    'No_of_Visitors': ['sum', 'count'], # 'count' here tells us the number of transaction line items
    'Unit_Price': ['mean', 'max', 'min']
}).reset_index()

summary_stats
```

```python
# Pre-Processing: Aggregating by Grouping and Park
def get_summary(df, park_name):
    park_df = df[df['Park'] == park_name].copy()
    summary = park_df.groupby('Grouping').agg({
        'No_of_Visitors': 'sum',
        'Total_Revenue': 'sum'
    }).reset_index()
    # Calculate weighted average yield per channel
    summary['Yield'] = summary['Total_Revenue'] / summary['No_of_Visitors']
    summary['Visitor_Pct'] = summary['No_of_Visitors'] / summary['No_of_Visitors'].sum()
    summary['Rev_Pct'] = summary['Total_Revenue'] / summary['Total_Revenue'].sum()
    return summary

sunway_summary = get_summary(df_channel_park, 'Sunway Lagoon')
lw_summary = get_summary(df_channel_park, 'Lost World')
```

### Yield Waterfall (List Price, Dilusion and Actual Yield)

```python
# PLOT: YIELD WATERFALL (List Price -> Dilution -> Actual)

parks = ['Sunway Lagoon', 'Lost World']
list_prices = {'Sunway Lagoon': 185.0, 'Lost World': 150.0}
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
for i, park in enumerate(parks):
    p_df = df_channel_park[df_channel_park['Park'] == park]

    # Actual Yield
    actual_yield = p_df['Total_Revenue'].sum() / p_df['No_of_Visitors'].sum()

    # List Price (Benchmark)
    list_p = list_prices[park]
    
    # Calculate Dilution (Impact of Unit_Price == 0)/Amount of Revenue Lost
    non_rev_ratio = p_df[p_df['Unit_Price'] == 0]['No_of_Visitors'].sum() / p_df['No_of_Visitors'].sum()
    dilution_impact = -list_p * non_rev_ratio
    # Amount of revenue lost due to dilution = List Price * % of visitors that are non-revenue
    
    # Calculate Pricing Discount (The remaining leak)
    discount_impact = (actual_yield - list_p) - dilution_impact
    # Discount given to the paid customer due to myKad, child/senior rates, OTA commissions, early bird promotions
    
    vals = [list_p, dilution_impact, discount_impact, actual_yield]
    labels = ['List Price', 'Non-Rev Dilution', 'Pricing Discount', 'Final Yield']
    steps = np.cumsum([0] + vals[:-1])
    
    colors = ['#4C72B0', '#C44E52', '#CCB974', '#55A868']
    for j in range(4):
        axes[i].bar(labels[j], vals[j], bottom=steps[j], color=colors[j])
        axes[i].text(j, steps[j] + vals[j]/2, f"{vals[j]:.1f}", ha='center', color='white', weight='bold')
    axes[i].set_title(f"Yield Waterfall: {park}")

plt.tight_layout()
plt.show()
```

The diagram above shows the breakdown of the list price VS the final yield obtained. Non-rev dilution is the expected potential revenue lost when entertaining the individuals with free admissions to the park. 

Sunway Lagoon suffered more from giving free admissions compared to Lost World. However, this can be compensated by ther higher ticket price, and only a small portion (RM 14.6) on average is given to individuals, leading to a final yield of RM 32 per admission. 

On the other hand, Lost World suffered less from giving free admissions, but giving out more pricing discount to the paying customers. In addition, they also charge a lower ticket price (using RM 150 for the day admission), causing the final yield to be lower RM 17 per admission at the end. 

Sunway Lagoon generates higher yield by charging a higher price and giving out smaller pricing discount, and it is enough to cover the large amount of free admissions given. 
Lost World is giving out double the discount, but only yielding half of the final yield, though having a lower revenue loss.

### Revenue Concentration (Lorenz Curve)

```python
plt.figure(figsize=(10, 6))
for p_summary, name, color in [(sunway_summary, 'Sunway Lagoon', 'orange'), (lw_summary, 'Lost World', 'blue')]:
    temp = p_summary.sort_values('Yield', ascending=False)
    cv = np.insert(np.cumsum(temp['Visitor_Pct']).values, 0, 0)
    cr = np.insert(np.cumsum(temp['Rev_Pct']).values, 0, 0)
    plt.plot(cv, cr, label=name, color=color, marker='o', markersize=4)
    
plt.plot([0, 1], [0, 1], '--', color='grey')
plt.title("Revenue Concentration: Visitors vs Revenue")
plt.xlabel("Cumulative % of No_of_Visitors")
plt.ylabel("Cumulative % of Total_Revenue")
plt.legend()
plt.show()
```

The Lorenz Curve shows the concentration of revenue within the visitors we have. This complements our figure 3, which shows our client mix.

The Lorenz Curve proves that Sunway Lagoon operates a High-Concentration Model, where under 25% of visitors generate nearly 95% of all entrance revenue. This creates a high dependency on 'Premium' segments like Walk-Ins. In contrast, Lost World’s revenue is more distributed, requiring 40% of its visitors to reach the same revenue threshold. While Sunway enjoys higher efficiency from its top-tier guests, Lost World’s curve suggests a more balanced reliance across multiple paying channels.

### Visitor Mix by Yield Tiers

```python
def get_tiers(p_summary):
    high = p_summary[p_summary['Yield'] >= 100]['Visitor_Pct'].sum()
    mid = p_summary[(p_summary['Yield'] > 0) & (p_summary['Yield'] < 100)]['Visitor_Pct'].sum()
    zero = p_summary[p_summary['Yield'] == 0]['Visitor_Pct'].sum()
    return [high, mid, zero]

comp_df = pd.DataFrame({
    'Sunway Lagoon': get_tiers(sunway_summary),
    'Lost World': get_tiers(lw_summary)
}, index=['High Yield (>RM100)', 'Mid Yield (RM1-100)', 'Zero Yield (RM0)']).T

comp_df.plot(kind='bar', stacked=True, color=['#55A868', '#CCB974', '#C44E52'])
plt.title("Visitor Mix (No_of_Visitors) by Yield Tiers")
plt.ylabel("Composition Percentage")
plt.show()
```

This can better supplement the Lorenz Curve, illustrating what types of visitors/passengers contribute to the revenue concentration of both Sunway Lagoon ad Lost World.
