# Transactions by Channel (Online, Walk-in, etc) and Park

In this particular EDA, we want to analyze the transactions by channel and park (Sunway Lagoon VS Lost World). We would want to see if there is any particular trend, eg, more individuals are paying using online in Sunway Lagoon, etc. 

## List of variables

1. Environment: Specify whether the environment is Sunway Lagoon or Lost World
- WHEN Environment = 1 THEN 'Sunway Lagoon'
- WHEN Environment = 2 THEN 'Lost World'

2. Transaction data
- Complete_QTY: Total Visitors
- OriginalAmount * Complete_QTY: Total Revenue

3. Park: Revenue Grouping (whether you are online, walk-in, bulk buying, etc)
- We should understand the different unique values available

4. TranDate: Transaction Date
- According to our EDA, we only focus on the data from 2023 to 2024 to train our model, skipping earlier years that is affected by COVID-19.

## Loading Dataset

```python
# Connect to the database
con = duckdb.connect("theme_park_data.db")

# Extract the necessary information for EDA
# Transactions data, channels (revenue grouping), and type of park/environment
# Focus from 2023 to 2024

df_channel_park = con.execute("""
SELECT
    TRIM(Park) as Grouping,
    CASE
        WHEN Environment = 1 THEN 'Sunway Lagoon'
        WHEN Environment = 2 THEN 'Lost World'
    END AS Park,
    OriginalAmount AS Unit_Price,
    Complete_QTY AS No_of_Visitors,
    (OriginalAmount * Complete_QTY) AS Total_Revenue
FROM theme_park_database
WHERE Environment IN (1, 2) 
    AND Grouping IS NOT NULL
    AND TranDate >= TIMESTAMP '2023-01-01 00:00:00'
    AND TranDate < TIMESTAMP '2025-01-01 00:00:00'
""").df()

df_channel_park.head()
```

### 🟡 Dirty Data Processing 

Hence, here are the steps I would do to temporary "correct" the dataset before heading into generating and comparing the transactions. 
1. Correcting the sign of the ticket price (I suspect that is a typo of negative sign, and I believe that there should be a separate section by the data regarding reimbursement column)
2. Correcting the sign of the number of visitors (because number of visitor is never negative).
3. Dropping all the number of visitors which is <1, to ensure our result is still interpretable and not having weird visitor number. 

```python
# 1. Correcting the sign of the ticket price (OriginalAmount)
# Using absolute value to ensure all prices are positive
df_channel_park['Unit_Price'] = df_channel_park['Unit_Price'].abs()

# 2. Correcting the sign of the number of visitors (Complete_QTY)
# Ensuring that negative visitor counts (likely reversal logs) are treated as positive volume
df_channel_park['No_of_Visitors'] = df_channel_park['No_of_Visitors'].abs()

# 3. Drop fractional visitors but keep whole numbers (including 0)
# This keeps 0.0, 1.0, 2.0 etc., and removes 0.5, 1.7, etc.
initial_count = len(df_channel_park)
df_channel_park = df_channel_park[df_channel_park['No_of_Visitors'] % 1 == 0]

# Recalculate total_revenue to ensure it matches the corrected signs
df_channel_park['Total_Revenue'] = df_channel_park['Unit_Price'] * df_channel_park['No_of_Visitors']

dropped_count = initial_count - len(df_channel_park)

print(f"Data Cleaning Complete:")
print(f"- Fixed negative signs for Prices and Visitor counts.")
print(f"- Dropped {dropped_count} rows with visitor counts < 1.")
print(f"- New dataframe shape: {df_channel_park.shape}")
print(f"Amount of data lost: {dropped_count} rows ({(dropped_count/initial_count)*100:.2f}%)")

print("### Summary Statistics by Park and Channel ###")
summary_stats = df_channel_park.groupby(['Park', 'Grouping']).agg({
    'Total_Revenue': ['sum', 'mean', 'std'],
    'No_of_Visitors': ['sum', 'count'], # 'count' here tells us the number of transaction line items
    'Unit_Price': ['mean', 'max', 'min']
}).reset_index()

summary_stats
```

## Visual Summary

```python
# Compute yield per visitors
# Using the specific columns: Total_Revenue / No_of_Visitors
df_channel_park['Yield_Calc'] = (df_channel_park['Total_Revenue']/df_channel_park['No_of_Visitors'])
```

```python
# Clean the grouping for vsibility
# Group Tours/Hotels/Presold together
target_names = ['Tours/Hotels/Presold', 'Tour / Hotel / Presold']

# Replace them with your preferred standard name
df_channel_park['Grouping'] = df_channel_park['Grouping'].replace(target_names, 'Tours/Hotels/Presold')

# Group Intercompany/Staff together
target_names = ['Inter-company', 'Staff']

# Replace them with your preferred standard name
df_channel_park['Grouping'] = df_channel_park['Grouping'].replace(target_names, 'Inter-company/Staff')

# Standard check: Clean any extra spaces just in case
df_channel_park['Grouping'] = df_channel_park['Grouping'].str.strip()
```

```python
# First plot
# 1. Bar charts showing the total number of visiotors VS revenue that is generated by different grouping for Sunway and Lost World respectively
fig, axes = plt.subplots(3, 1, figsize=(16, 12))

# Subplot A: Total Revenue
sns.barplot(data=df_channel_park, x='Grouping', y='Total_Revenue', hue='Park', ax=axes[0])
axes[0].set_title('Total Revenue Generated by Grouping and Park', fontsize=15)
axes[0].set_ylabel('Total Revenue (RM)')
axes[0].tick_params(axis='x', rotation=45)

# Subplot B: Total Number of Visitors
sns.barplot(data=df_channel_park, x='Grouping', y='No_of_Visitors', hue='Park', ax=axes[1])
axes[1].set_title('Total Number of Visitors by Grouping and Park', fontsize=15)
axes[1].set_ylabel('Total Visitors')
axes[1].tick_params(axis='x', rotation=45)

# Subplot C: Yield per Visitor
sns.barplot(data=df_channel_park, x='Grouping', y='Yield_Calc', hue='Park', ax=axes[2])
axes[2].set_title('Revenue/Visitors by Grouping and Park', fontsize=15)
axes[2].set_ylabel('Revenue/Visitors')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 2. Overall Yield per Visitor: Comparison between the 2 Parks

# We calculate this at the aggregate level for the whole park
park_overall = df_channel_park.groupby('Park').agg(
    total_rev=('Total_Revenue', 'sum'),
    total_vis=('No_of_Visitors', 'sum')
).reset_index()
park_overall['overall_yield'] = park_overall['total_rev'] / park_overall['total_vis']

plt.figure(figsize=(8, 6))
sns.barplot(data=park_overall, x='Park', y='overall_yield', palette='viridis')
plt.title('Overall Yield per Visitor by Park', fontsize=14)
plt.ylabel('Average Revenue per Head (RM)')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

- Average Lost World Ticket Price: RM 95 - RM 150
- Average Sunway Lagoon Ticket Price: RM 185 (for adults), RM 150 for children and adults

At first glance, there is a RM 35 difference in the walk-in ticket prices (RM 185 for Sunway vs. RM 150 for Lost World). However, the data shows the actual difference in average revenue per head is only RM 15.5.

This indicates that Sunway Lagoon has a higher 'dilution rate.' While it commands a more premium gate price, it also carries a significantly larger volume of non-revenue-generating visitors (such as large-scale corporate events, B2B night events, and complimentary passes) compared to Lost World.

Conclusion: We can infer that while both parks use complimentary admissions to drive volume, Sunway Lagoon’s business model relies more heavily on these '0-revenue' segments to maintain high foot traffic, whereas Lost World’s yield is more closely aligned with its ticket pricing, suggesting a more direct-to-consumer revenue model.

We will use the visitor mix chart to justify if Sunway Lagoon's traffic is driven more by those of free admissions. (O revenue), which includes:

```python
# Visitor Mix Chart: Paid vs. Non-Revenue (%)

# Define 'Status' based on whether the transaction generated revenue
df_channel_park['Status'] = np.where(df_channel_park['Yield_Calc'] == 0.0, 'Non-Revenue', 'Paid')

# Aggregate the volume (No_of_Visitors) by Status
mix = df_channel_park.groupby(['Park', 'Status'])['No_of_Visitors'].sum().unstack().fillna(0)

# Convert to percentages
mix_perc = mix.div(mix.sum(axis=1), axis=0) * 100

ax = mix_perc.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#ff9999', '#66b3ff'])
plt.title('Visitor Mix (% Paid vs. Non-Revenue Capacity)', fontsize=14)
plt.ylabel('Percentage of Total Visitors (%)')
plt.legend(title='Guest Status', bbox_to_anchor=(1, 1))

# Add percentage labels directly on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    if height > 0:
        ax.text(p.get_x() + width/2, p.get_y() + height/2, f'{height:.1f}%', 
                ha='center', va='center', color='white', fontweight='bold')

plt.xticks(rotation=0)
plt.show()
```

The Visitor Mix chart confirms that both parks operate on a high-volume, low-yield model, with approximately 70-75% of visitors generating zero direct revenue at the door.
